```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from sklearn.svm import SVC
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Load Fisher's Iris Dataset

irisData = load_iris ()
X = irisData.data
Y = irisData.target

# Testing sizes 1/4, 1/2, 3/4, 8/10

testSizes = [0.25, 0.5, 0.75, 0.8]
randomizerSeeds = numpy.arange(1, 21, 1)

# 5 different classification training models

Models = { 
  'Support Vector SVC' : SVC(),
  'Naive Bayes' : GaussianNB(),
  'Decision Tree' : DecisionTreeClassifier(),
  'AdaBoost' : AdaBoostClassifier(),
  'Quadratic Discriminant' : QuadraticDiscriminantAnalysis()
}    

# Iterate every model for every random seed and every test size  

DF = pandas.DataFrame()
roundNum = 3
modelName = []
testSizeUsed=[]
randomizerSeedUsed=[]
modelAccuracyObtained = []

for name, model in Models.items():
    for randomSeed in randomizerSeeds:
        for testSize in testSizes:
            X_train , X_test , Y_train , Y_test = train_test_split (X, Y, test_size = testSize , random_state = randomSeed)
            model.fit(X_train,Y_train);
            Y_hat = model.predict(X_test)
            acc = numpy.sum(Y_hat == Y_test)/len(Y_hat)
            modelName.append(name)
            modelAccuracyObtained.append(acc);
            testSizeUsed.append(testSize)
            randomizerSeedUsed.append(randomSeed)

# Assign the results to a Data Frame

DF=DF.assign(**{"Model Name" : modelName})
DF=DF.assign(**{"Test Size" : testSizeUsed})
DF=DF.assign(**{"Random Seed" : randomizerSeedUsed})
DF=DF.assign(**{"Accuracy" : modelAccuracyObtained})
```

```
# Calculate the mAP for every model and every Test Size

SVCModel = DF.loc[ (DF['Model Name'] == 'Support Vector SVC') ]
SVCModelOne = SVCModel.loc[ (SVCModel['Test Size'] == 0.25) ]
SVCModelTwo = SVCModel.loc[ (SVCModel['Test Size'] == 0.5) ]
SVCModelThree = SVCModel.loc[ (SVCModel['Test Size'] == 0.75) ]
SVCModelFour = SVCModel.loc[ (SVCModel['Test Size'] == 0.8) ]
SVCMean = SVCModel["Accuracy"].mean()

NaiveBayesModel = DF.loc[ (DF['Model Name'] == 'Naive Bayes') ]
NaiveBayesModelOne = NaiveBayesModel.loc[ (NaiveBayesModel['Test Size'] == 0.25) ]
NaiveBayesModelTwo = NaiveBayesModel.loc[ (NaiveBayesModel['Test Size'] == 0.5) ]
NaiveBayesModelThree = NaiveBayesModel.loc[ (NaiveBayesModel['Test Size'] == 0.75) ]
NaiveBayesModelFour = NaiveBayesModel.loc[ (NaiveBayesModel['Test Size'] == 0.8) ]
NaiveBayesMean = NaiveBayesModel["Accuracy"].mean()

DecisionTreeModel = DF.loc[ (DF['Model Name'] == 'Decision Tree') ]
DecisionTreeModelOne = DecisionTreeModel.loc[ (DecisionTreeModel['Test Size'] == 0.25) ]
DecisionTreeModelTwo = DecisionTreeModel.loc[ (DecisionTreeModel['Test Size'] == 0.5) ]
DecisionTreeModelThree = DecisionTreeModel.loc[ (DecisionTreeModel['Test Size'] == 0.75) ]
DecisionTreeModelFour = DecisionTreeModel.loc[ (DecisionTreeModel['Test Size'] == 0.8) ]
DecisionTreeMean = DecisionTreeModel["Accuracy"].mean()

AdaBoostModel = DF.loc[ (DF['Model Name'] == 'AdaBoost') ]
AdaBoostModelOne = AdaBoostModel.loc[ (AdaBoostModel['Test Size'] == 0.25) ]
AdaBoostModelTwo = AdaBoostModel.loc[ (AdaBoostModel['Test Size'] == 0.5) ]
AdaBoostModelThree = AdaBoostModel.loc[ (AdaBoostModel['Test Size'] == 0.75) ]
AdaBoostModelFour = AdaBoostModel.loc[ (AdaBoostModel['Test Size'] == 0.8) ]
AdaBoostMean = AdaBoostModel["Accuracy"].mean()

QuadraticDiscrModel = DF.loc[ (DF['Model Name'] == 'Quadratic Discriminant') ]
QuadraticDiscrModelOne = QuadraticDiscrModel.loc[ (QuadraticDiscrModel['Test Size'] == 0.25) ]
QuadraticDiscrModelTwo = QuadraticDiscrModel.loc[ (QuadraticDiscrModel['Test Size'] == 0.5) ]
QuadraticDiscrModelThree = QuadraticDiscrModel.loc[ (QuadraticDiscrModel['Test Size'] == 0.75) ]
QuadraticDiscrModelFour = QuadraticDiscrModel.loc[ (QuadraticDiscrModel['Test Size'] == 0.8) ]
QuadraticDiscrMean = QuadraticDiscrModel["Accuracy"].mean()

# mAP Data Frame

ModelsVector = ['Support Vector SVC', 'Naive Bayes', 'Decision Tree', 'AdaBoost', 'Quadratic Discriminant']

mAPOne = [round(SVCModelOne["Accuracy"].mean(), roundNum), 
          round(NaiveBayesModelOne["Accuracy"].mean(), roundNum), 
          round(DecisionTreeModelOne["Accuracy"].mean(), roundNum), 
          round(AdaBoostModelOne["Accuracy"].mean(), roundNum), 
          round(QuadraticDiscrModelOne["Accuracy"].mean(), roundNum)
]

mAPTwo = [round(SVCModelTwo["Accuracy"].mean(), roundNum), 
          round(NaiveBayesModelTwo["Accuracy"].mean(), roundNum), 
          round(DecisionTreeModelTwo["Accuracy"].mean(), roundNum), 
          round(AdaBoostModelTwo["Accuracy"].mean(), roundNum), 
          round(QuadraticDiscrModelTwo["Accuracy"].mean(), roundNum)
]

mAPThree = [round(SVCModelThree["Accuracy"].mean(), roundNum), 
            round(NaiveBayesModelThree["Accuracy"].mean(), roundNum), 
            round(DecisionTreeModelThree["Accuracy"].mean(), roundNum), 
            round(AdaBoostModelThree["Accuracy"].mean(), roundNum), 
            round(QuadraticDiscrModelThree["Accuracy"].mean(), roundNum)
]

mAPFour = [round(SVCModelFour["Accuracy"].mean(), roundNum), 
           round(NaiveBayesModelFour["Accuracy"].mean(), roundNum), 
           round(DecisionTreeModelFour["Accuracy"].mean(), roundNum), 
           round(AdaBoostModelFour["Accuracy"].mean(), roundNum), 
           round(QuadraticDiscrModelFour["Accuracy"].mean(), roundNum)
]

mAPFive = [round(SVCMean, roundNum), 
           round(NaiveBayesMean, roundNum), 
           round(DecisionTreeMean, roundNum), 
           round(AdaBoostMean, roundNum), 
           round(QuadraticDiscrMean, roundNum)
]

DFMAP = pandas.DataFrame()
DFMAP = DFMAP.assign(**{"Model Name" : ModelsVector})
DFMAP = DFMAP.assign(**{"mAP 1/4 TS" : mAPOne})
DFMAP = DFMAP.assign(**{"mAP 1/2 TS" : mAPTwo})
DFMAP = DFMAP.assign(**{"mAP 3/4 TS" : mAPThree})
DFMAP = DFMAP.assign(**{"mAP 8/10 TS" : mAPFour})
DFMAP = DFMAP.assign(**{"mAP AVG TS" : mAPFive})
```
