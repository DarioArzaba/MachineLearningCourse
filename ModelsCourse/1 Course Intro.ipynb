##Vector and Matrix##

**Vector** 

Mathematical structure that represents a variable in lineal space.

$ v = {a_0, a_1, ... , a_i, ... a_n} \forall a_i \in \mathbb{R} \therefore v \in \mathbb{R}^n $

**Matrix**

Structure mede up of vectors that uses a structural order to represent the dimention of the vector, in other words they represent vectors in a structured way.

$ v = {v_0, v_1, ... , v_i, ... v_m} \forall v_i \in \mathbb{R}^n \therefore v \in \mathbb{R}^{n \times m} $

```python
vector = numpy.asarray ([1 ,2 ,3])
print("This is a vector:")
print(vector)
print()

matrix = numpy.asarray ([ vector , [3, 4, 5]])
print("This is a matrix:")
print(matrix)
print()

randMatrix = numpy.random.randint(0, 9, (100,30))
print("Matrix with random int (0-9) of 100 rows and 30 columns:")
print(randMatrix.shape)
print()

randMatrix = []
for i in range(0, 100):
    randMatrix.append(numpy.random.randint(0, 10, 30))
randMatrix = numpy.asarray(randMatrix)
print("Matrix of 100x30 created by appending vectors with a for cycle:")
print(randMatrix.shape)
print()

# Use a CSV file to create a matrix:
# fileMatrix = pandas.read_csv('data.csv', header=None).to_numpy()
```

##Transformation##

**Transformation**

Matrix product that allows for a lineal maping of values from $\mathbb{R}^p$ into $\mathbb{R}^d$. 

$T(v) = (A)(v) : v \in \mathbb{R}^n $

In general there is always a mapping of (n) vectors in matrix (X) that originate from the transformation of matrix (A).

$Y_{n\times d} = X_{n\times m} A_{m \times p}$

```python
print("Matrix of 100x30 transformed into a matrix of 100x15.")
print("By using the resize method:")
matrix = numpy.random.rand(100,30)
print(matrix.shape)
matrix.resize((100, 15))
print(matrix.shape)
print()

print("Matrix of 100x30 transformed into a matrix of 100x15.")
print("By using a matrix product:")
matrixX = numpy.random.rand(100,30)
matrixA = numpy.random.rand(30,15)
matrixY = numpy.matmul(matrixX, matrixA)
print(matrixY.shape)
```

##Error##

**Error**

Difference from the desired value and the actual value of a model. If $y$ is the desired value and $\hat{y} = f(x)$ is the actual value then $e = \hat{y} - y$.

```python
print("Evaluate the error between two random vectors.")
print("Each vector has 100 datapoints:")
Y = numpy.random.rand(100, 1)
Y_hat = numpy.random.rand(100,1)
errorMatrix = []
for (y, y_hat) in zip(Y, Y_hat):
    e = abs(y_hat - y)
    errorMatrix.append(e)
errorMatrix = numpy.asarray(errorMatrix)
# Final error value is the average of every error datapoint
print(numpy.average(errorMatrix))
```

##Generalization and Training##

**Generalization**

Capacity of a model of predicting a stable output from variable inputs.

**Training**

Process of providing the model inputs until the model aquires the skill needed to achieve the desired output. Each step in the training is called an epoch, note that iterations are usually greater than the number of epochs.

1. Underfitting: The model doesn't learn the input patterns.
2. Overfittig: The model memorizes the specific input used instead of learning its patterns.

##Normalization and Parameters##

**Normalization**

Process of making the inputs invariant to changes in the unit scale used.

**Parameters**

Values that control the behavior of a model.

```python
print("Plot the sin(X) model by changing alpha and beta parameters:")
plt.figure(figsize = [8, 6])
x = numpy.arange(-3.1416, 3.1416, 0.1)
a_1 = 1
a_2 = 2
b_1 = 1
b_2 = 2
y_1 = a_1 * numpy.sin(x * b_1)
y_2 = a_2 * numpy.sin(x * b_1)
y_3 = a_1 * numpy.sin(x * b_2)
y_4 = a_2 * numpy.sin(x * b_2)
plt.plot(x, y_1, label = 'sin (x)')
plt.plot(x, y_2, label = '2 sin (x)')
plt.plot(x, y_3, label = 'sin (2x)')
plt.plot(x, y_4, label = '2 sin (2x)')
plt.title(r'$\alpha ~ \sin (\beta x)$')
plt.xlabel('x')
plt.legend(loc = 'upper right')
plt.show()
```

##Metric##

**Metric**

Scalar representing the generalization capacity of a model, its range and interpretation can vary from system to system. The most basic metric is simply counting the times the model has produces the desired output and using it as a percentage of the total output.

```python
print("Two random int vectors (0-9) of 100 points.")
print("Metric in % representing the success rate of the output:")
Y = numpy.random.randint(0, 10, 100)
Y_hat = numpy.random.randint(0, 10, 100)
acc = 0
for (y, y_hat) in zip(Y, Y_hat):
    acc += (y == y_hat)
acc /= Y.size
print(acc)
# Since there are 10 posible digits and the distribution
# is random we expect a 10% or 0.1 success rate.
```

##Ground Truth and Confusion Matrix##

**Ground Truth**

Group of tags used to associate each input to its correct output.

**Confusion Matrix**

Metric used to understand where the model is having problems when making discrete predictions. Each input is assigned to each output for every cell in the table, then we count how many times does the model asign a particular output to an input and we do this for every output. 

Note that for a classification algorithm every input-output assigment is a wrong except for the diagonal of the matrix (correct input to output pairs).

```python
print("Two random int vectors (0-9) of 100 points.")
print("Generate the confusion matrix of the model:")
Y = numpy.random.randint(0, 10, 100)
Y_hat = numpy.random.randint(0, 10, 100)
acc = 0
confusionMatrix = numpy.zeros((10,10))
for y, y_hat in zip(Y, Y_hat):
    confusionMatrix[y][y_hat] += 1
confusionMatrixDiagonal = numpy.diag(confusionMatrix)
acc = confusionMatrixDiagonal.sum() / Y.size
print(confusionMatrix)
print('Success rate: ' + str(acc))
```

##Class Balance##

**Class Balance**

For a classification algorithm we need to have the same number of inputs for each class since class unbalances will cause training to ignore wrong predictions for classes with a small input dataset.

```python
print("Random int vector (0-9) of 100 points.")
print("Graph class balance for each input (digit):")
Y = numpy.random.randint(0, 10, 100)
plt.hist(Y, bins = 10)
countInputsPerClass = numpy.unique(Y, return_counts = True)
countInputsPerClass = numpy.asarray(countInputsPerClass)
print(countInputsPerClass)
```

##Characteristic Vector##

**Characteristic Vector**

Vector that represents a real object as a mathematical entity.

```python
header = ['Age', 'Weight', 'Height']
data = [ [1.2, 12, 0.93], [2.0,14,0.80], [3.0,15,0.70] ]

def IMC(weight, height):
    return weight/(height**2)

with open('CharVector.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)
DF = pandas.read_csv('CharVector.csv')

IMCVector = []
for (weight, height) in zip(DF.Weight, DF.Height):
    IMCDataPoint = IMC(weight, height)
    IMCVector.append(IMCDataPoint)
DF = DF.assign(IMC = IMCVector)

print(DF)
```

##Descriptive Vector##

**Descriptive Vector**

```python
DF["IMC2"] = DF.apply(lambda x: (x.Weight/x.Height**2), axis=1)
DF
```
