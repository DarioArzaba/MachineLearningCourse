```
import pandas
import glob
import numpy
import re
import csv

File = glob.glob("./OriginalDataset/*/*")
DF = pandas.DataFrame()
DF = DF.assign(File = File)
DF["Class"] = DF.apply(lambda x: numpy.take( re.findall(r'Dataset/C(\d+)', x.File) ,0), axis=1)
DF.to_csv('DATASET.csv')
print(DF)
```
