```python
import numpy 
import pandas
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn . model_selection import train_test_split

data = load_iris()
X = data.data
Y = data.target

Models = { 
    'kNN -3' : KNeighborsClassifier ( n_neighbors = 3) ,
    'kNN -5' : KNeighborsClassifier ( n_neighbors = 5) };

TS = [0.4,0.5,0.6,0.7,0.8,0.9]
RS = [10,42,125,999]
NM = []
ACC = [];
ARS=[]
ATS=[]
DF=pandas.DataFrame()

for name, model in Models.items():
  for rs in RS:
    for ts in TS:
      X_train , X_test , Y_train , Y_test = train_test_split (X, Y, test_size =ts , random_state = rs);
      model.fit(X_train,Y_train);
      Y_hat = model.predict(X_test)
      acc = numpy.sum(Y_hat == Y_test)/len(Y_hat)
      # NM
      NM.append(name)
      # KNN
      ACC.append(acc);
      #TS
      ATS.append(ts)
      #RS
      ARS.append(rs)

DF=DF.assign(NM=NM)
DF=DF.assign(RS=ARS)
DF=DF.assign(TS=ATS)
DF=DF.assign(ACC=ACC)
print(DF)

DF = DF.nlargest(10,['ACC'])
print(DF)
```

```python
import matplotlib.pyplot as plt
import numpy
g = 0.001; # eta

D = 2;

X = numpy.random.rand(100,D);
Y = numpy.random.rand(100,1);

epochs = 30;

W = numpy.random.rand(D ,1);
B = 1- numpy.random.rand(1 ,1);

Loss = []

for _ in range( epochs ):
  loss = 0;
  for (x,y) in zip(X,Y):
    y_hat = numpy.matmul(x, W)+B;
    e = y_hat - y;
    W -= g * e;
    loss += e;
  loss /= len(X);
  Loss.append(loss[0][0])

plt.plot(Loss)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show
```

```python
import numpy
import pandas
from sklearn.linear_model import Ridge
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from scipy import stats
X, Y = load_diabetes (return_X_y =True)
X_train , X_test , Y_train , Y_test = train_test_split (X, Y, test_size =0.2);
R = [];
A = [];


for alpha in numpy. linspace (0.1 ,0.99 ,10):
    Model = Ridge(alpha = alpha )
    Model .fit(X_train , Y_train )
    Y_hat = Model. predict ( X_test )
    rho, _ = stats. spearmanr (Y_hat , Y_test )
    A.append(alpha)
    R.append(rho)

DF = pandas.DataFrame();
DF = DF.assign(Lambda = A);
DF = DF.assign(Rho = R);

print(DF)
```

```python
import numpy
import pandas
from sklearn.linear_model import Ridge, Lasso, SGDRegressor, LassoCV
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from scipy import stats
X, Y = load_diabetes (return_X_y =True)

Models = { 
    'Rige' : Ridge(alpha=0.3) ,
    'Lasso' : Lasso(alpha=0.3) ,
    'LassoCV': LassoCV(),
    'SGDRegressor': SGDRegressor()};

TS = [0.98, 0.7, 0.8]
RS = [37, 46, 33]
NM = []
RHO = [];
ARS=[]
ATS=[]
DF=pandas.DataFrame()

for name, model in Models.items():
  for rs in RS:
    for ts in TS:
      X_train , X_test , Y_train , Y_test = train_test_split (X, Y, test_size =ts , random_state = rs);
      model.fit(X_train,Y_train);
      Y_hat = model.predict(X_test)
      rho, _ = stats. spearmanr (Y_hat , Y_test )
      # NM
      NM.append(name)
      # RHO
      RHO.append(rho);
      #TS
      ATS.append(ts)
      #RS
      ARS.append(rs)

DF=DF.assign(NM=NM)
DF=DF.assign(RS=ARS)
DF=DF.assign(TS=ATS)
DF=DF.assign(RHO=RHO)
print(DF)

DF = DF.nlargest(10,['RHO'])
print(DF)
```

## ANOVA con k constantes

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest, chi2
import numpy
import pandas
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestCentroid
from sklearn import tree
from sklearn.neural_network import MLPClassifier

# cargando el dataset de pacientes con tumores
data = load_breast_cancer();

# separar las etiquetas y los datos 
X, Y = load_breast_cancer(return_X_y =True)


Models = { 
    'BayesNaive' : GaussianNB() ,
    'SVM' : svm.SVC(),
    'KNN': NearestCentroid(),
    'Tree': tree.DecisionTreeClassifier(criterion = 'entropy'),
    'ANN': MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1),
    };

NM = [] # Nombre del model
K = [];# k de ANOVA
ACC=[] # precision 

DF=pandas.DataFrame()

for name, model in Models.items():
  for k in range(1,20):
    X_hat = SelectKBest (chi2 , k = k).fit_transform (X,Y);
    X_train, X_test, Y_train, Y_test = train_test_split(X_hat, Y, test_size = 0.8, random_state=0)
    model.fit(X_train,Y_train);
    Y_hat = model.predict(X_test)
    acc = numpy.mean(Y_hat == Y_test);
    # NM
    NM.append(name)
    # k
    K.append(k);
    # acc
    ACC.append(acc)

DF=DF.assign(NM=NM)
DF=DF.assign(K=K)
DF=DF.assign(ACC=ACC)
print(DF)

DF = DF.nlargest(10,['ACC'])
print(DF)
```

## ANOVA con porcentajes

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest, chi2, SelectPercentile
import numpy
import pandas
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestCentroid
from sklearn import tree
from sklearn.neural_network import MLPClassifier

# cargando el dataset de pacientes con tumores
data = load_breast_cancer();

# separar las etiquetas y los datos 
X, Y = load_breast_cancer(return_X_y =True)


Models = { 
    'BayesNaive' : GaussianNB() ,
    'SVM' : svm.SVC(),
    'KNN': NearestCentroid(),
    'Tree': tree.DecisionTreeClassifier(),
    'ANN': MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1),
    };

NM = [] # Nombre del model
PER = [];# Percetile de ANOVA
ACC=[] # precision 

DF=pandas.DataFrame()

for name, model in Models.items():
  for percentile in [5,10,20,30,40,50,60]:
    X_hat = SelectPercentile (chi2 , percentile = percentile).fit_transform (X, Y)
    X_train, X_test, Y_train, Y_test = train_test_split(X_hat, Y, test_size = 0.8, random_state=0)
    model.fit(X_train,Y_train);
    Y_hat = model.predict(X_test)
    acc = numpy.mean(Y_hat == Y_test);
    # NM
    NM.append(name)
    # k
    PER.append(percentile);
    # acc
    ACC.append(acc)

DF=DF.assign(NM=NM)
DF=DF.assign(PER=PER)
DF=DF.assign(ACC=ACC)
print(DF)

DF = DF.nlargest(10,['ACC'])
print(DF)
```
