# Modelos Mixtos

Este es el codigo para entrenar un modelo que hacen predicciones probabilisticas

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import numpy

# Leer el dataset de Fisher
X, Y = load_iris ( return_X_y =True)

# Proyectar en 2-Dimensiones
pca = PCA( n_components =2)
pca.fit(X)

# El dataset proyectado
X_hat = pca.transform (X)

# Visualizando el dataset
for k, lbl in enumerate(['Setosa', 'Versicolor','Virginica']):
  plt.scatter (X_hat[Y == k,0], X_hat [Y == k,1],label = lbl);

plt.title("Fisher Iris")
plt.legend (loc="upper left");

plt.show();

# Entrenar el modelo
Model = GaussianMixture(n_components = 3, random_state = 0).fit(X_hat);

Y_hat = Model.predict(X_hat)

Y_hat[Y_hat == 2] = 3;
Y_hat[Y_hat == 1] = 2;
Y_hat[Y_hat == 3] = 1;

print('acc = %0.4f' %(numpy.sum( Y_hat == Y )/len(Y) ) )
```
